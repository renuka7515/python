```python
import random

# Function to roll the dice
def roll_dice():
    return random.randint(1, 6)

# Function to play the snake and ladder game
def play_game():
    player_positions = [0, 0]
    current_player = 0

    while True:
        # Switch to the next player
        current_player = 1 if current_player == 0 else 0
        player_name = "Player " + str(current_player + 1)

        # Wait for the user to press Enter
        input(player_name + ", press Enter to roll the dice...")

        # Roll the dice
        roll = roll_dice()
        print(player_name + " rolled a " + str(roll))

        if roll != 6 and player_positions[current_player] == 0:
            print(player_name + " needs a 6 to start the game.")
        else:
            player_positions[current_player] += roll

            # Define the snake and ladder positions
            snakes = {16: 6, 47: 26, 49: 11, 56: 53, 62: 19, 64: 60, 87: 24, 93: 73, 95: 75, 98: 78}

            ladders = {1: 38, 4: 14, 9: 31, 21: 42, 28: 84, 36: 44, 51: 67, 71: 91, 80: 100}

            if player_positions[current_player] in snakes:
                print("Oops! " + player_name + " got bitten by a snake.")
                player_positions[current_player] = snakes[player_positions[current_player]]
            elif player_positions[current_player] in ladders:
                print("Great! " + player_name + " climbed a ladder.")
                player_positions[current_player] = ladders[player_positions[current_player]]

            print(player_name + " is now at position " + str(player_positions[current_player]))

            if player_positions[current_player] >= 100:
                print("Congratulations! " + player_name + " wins!")
                break

play_game()
```
